```python
import torch
import torch.nn as nn
import torch.optim as optim
import logging
from tqdm import tqdm
from collections import defaultdict

class TextPreProcessor:
    def __init__(self, min_freq=1):
        self.min_freq = min_freq

    def build_vocab(self, corpus):
        """
        Build the vocabulary based on the given corpus.

        :param corpus: List of sentences (each sentence is a list of words).
        :return: A dictionary mapping words to indices (word_to_ix) and indices to words (ix_to_word).
        """
        word_freq = defaultdict(int)
        for sentence in corpus:
            for word in sentence:
                word_freq[word] += 1

        # Filter out words below the minimum frequency threshold
        self.vocab = {word: i for i, (word, freq) in enumerate(word_freq.items()) if freq >= self.min_freq}
        self.vocab["<unk>"] = len(self.vocab)  # Add an unknown token for rare words

        # Mapping from word to index and vice versa
        word_to_ix = {word: idx for idx, word in enumerate(self.vocab.keys())}
        ix_to_word = {idx: word for word, idx in word_to_ix.items()}
        return word_to_ix, ix_to_word

class CBOW(nn.Module):
    def __init__(self, vocab_size, embedding_dim=100):
        super(CBOW, self).__init__()
        self.num_epochs = 100
        self.context_size = 20
        self.embedding_dim = embedding_dim
        self.learning_rate = 0.001
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

        # Embedding and network layers
        self.embeddings = nn.Embedding(vocab_size, self.embedding_dim)
        self.linear1 = nn.Linear(self.embedding_dim, 128)
        self.activation_function1 = nn.ReLU()
        self.linear2 = nn.Linear(128, vocab_size)
        self.activation_function2 = nn.LogSoftmax(dim=-1)

        self.model_path = "cbow_model.pth"

    def make_context_vector(self, context, word_to_ix) -> torch.LongTensor:
        """
        For each word in the context, find its corresponding index in the word_to_ix mapping.

        :param context: List of words representing the context.
        :param word_to_ix: Dictionary mapping words to their corresponding indices.
        :return: torch.LongTensor containing the indices of the context words.
        """
        idxs = [word_to_ix.get(w, word_to_ix["<unk>"]) for w in context]
        tensor = torch.LongTensor(idxs)
        return tensor

    def forward(self, inputs):
        """
        Forward pass of the CBOW model.

        :param inputs: Tensor of input word indices.
        :return: Log probabilities for each word in the vocabulary.
        """
        embeds = self.embeddings(inputs).mean(dim=0).view(1, -1)
        out = self.activation_function1(self.linear1(embeds))
        out = self.activation_function2(self.linear2(out))
        return out

    def build_training_data(self, corpus, context_size):
        """
        Prepare training data for the CBOW model.

        :param corpus: List of sentences (each sentence is a list of words).
        :param context_size: Number of words to consider before and after the target word.
        :return: List of (context, target) pairs.
        """
        data = []
        for sentence in corpus:
            for i, word in enumerate(sentence):
                # Get the context words around the target word
                start = max(0, i - context_size)
                end = min(len(sentence), i + context_size + 1)
                context = [sentence[j] for j in range(start, end) if j != i]
                data.append((context, word))
        return data

    def train_model(self, data, word_to_ix):
        # Loss and optimizer
        self.to(self.device)
        optimizer = optim.Adam(self.parameters(), lr=self.learning_rate)
        loss_function = nn.NLLLoss()

        logging.warning('Starting forward pass')
        for epoch in tqdm(range(self.num_epochs)):
            # Track total loss for the epoch
            total_loss = 0

            # Training loop for each (context, target) pair in the data
            for context, target in data:
                context_vector = self.make_context_vector(context, word_to_ix).to(self.device)

                # Forward pass
                log_probs = self(context_vector)

                # Calculate the loss
                loss = loss_function(log_probs, torch.tensor([word_to_ix.get(target, word_to_ix["<unk>"])], device=self.device))
                total_loss += loss.item()

                # Backward pass and optimization
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()

            # Log metrics to monitor loss
            logging.warning(f"End of epoch {epoch+1} | Loss: {total_loss:.3f}")

        # Save the trained model
        torch.save(self.state_dict(), self.model_path)
        logging.warning(f'Model saved to {self.model_path}')

    def predict(self, context, word_to_ix, ix_to_word, top_k=1):
        """
        Make predictions based on the provided context.

        :param context: List of context words.
        :param word_to_ix: Dictionary mapping words to their indices.
        :param ix_to_word: Dictionary mapping indices to words.
        :param top_k: Number of top predictions to return.
        :return: List of top-k predicted words.
        """
        # Convert context to a context vector
        context_vector = self.make_context_vector(context, word_to_ix).to(self.device)

        # Forward pass through the model
        with torch.no_grad():
            log_probs = self(context_vector)

        # Get the top-k predictions
        _, predicted_indices = torch.topk(log_probs, top_k, dim=-1)
        predicted_words = [ix_to_word[idx.item()] for idx in predicted_indices[0]]

        return predicted_words
```
